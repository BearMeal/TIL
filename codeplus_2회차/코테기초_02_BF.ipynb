```python
# 일곱난재이
# 머지소트한후 nCr해준다
# 근데 이때 7개의 합 =100을 찾느냐
# 9개의 합 -x = 100일때 합이 x인 두개의 조합을 찾느냐 차이? 

nlist = [int(input()) for _ in range(9)]
# nlist = [20,7,23,19,10,15,25,8,13]

def merge_sort(nlist):
    if len(nlist) <=1:
        return
    
    mid = (len(nlist) // 2)
    left = nlist[:mid]
    right = nlist[mid:]
    
    merge_sort(left)
    merge_sort(right)
    
    i=j=k=0
    while i < len(left) and j<len(right):
        if left[i] < right[j]:
            nlist[k] = left[i]
            i+=1
        else:
            nlist[k] = right[j]
            j+=1
        k+=1
            
    while i<len(left):
        nlist[k] = left[i]
        i+=1
        k+=1
        
    while j<len(right):
        nlist[k] = right[j]
        j+=1
        k+=1
        
merge_sort(nlist)
# print(nlist)

total = sum(nlist)
ans = []
def nCr(n,r,temp=[]):
    if len(ans) >=7: 
        return
    if n == len(nlist):
        if r == len(temp):
            if sum(temp) == 100:
                for i in temp:
                    ans.append(i)
        return
    
    temp.append(nlist[n])
    nCr(n+1,r,temp)
    temp.pop()
    nCr(n+1,r,temp)
nCr(0,7)

for i in ans:
    print(i)
```

```python
# 사탕게임

n = int(input())
nlist = [list(input()) for _ in range(n)]

# 오른쪽과 아래쪽만 바꿔보면서 전체를 조회한다
def check_maxlen(nlist):
    global maxlen
    # 행길이
    for i in nlist:
        cnt = 1
        ex = ''
        for j in i:
            if ex == j:
                cnt +=1
                ex = j
            else:
                ex = j
                cnt = 1
            if cnt > maxlen:
                maxlen = cnt
        
    #열길이
    for j in range(n):
        cnt =1
        ex = ''
        for i in range(n):
            if ex == nlist[i][j]:
                cnt +=1
                ex = nlist[i][j]
            else:
                ex = nlist[i][j]
                cnt = 1
            if cnt > maxlen:
                maxlen = cnt

maxlen = 1
check_maxlen(nlist)

for i in range(n):
    for j in range(n):
        if j+1 <n: #오른쪽
            nlist[i][j], nlist[i][j+1] = nlist[i][j+1], nlist[i][j]
            check_maxlen(nlist)
            nlist[i][j], nlist[i][j+1] = nlist[i][j+1], nlist[i][j]
        if i+1 <n: #아래
            nlist[i][j], nlist[i+1][j] = nlist[i+1][j], nlist[i][j]
            check_maxlen(nlist)
            nlist[i][j], nlist[i+1][j] = nlist[i+1][j], nlist[i][j]
            
print(maxlen)
        
```
