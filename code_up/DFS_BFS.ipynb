```python
# 숫자 등고선
n = int(input())
x,y = map(int,input().split())
graph = [[0]*(n+1) for i in range(n+1)]

# 왼쪽위
def dfs(graph,x,y,start,n):
    graph[x][y] = start
    for i in graph:
        print(i)
    print()
    #위
    if x>0:
        dfs(graph,x-1,y,start+1,n)
    # 왼
    if y>0:
        dfs(graph,x,y-1,start+1,n)
  
# 오른쪽아래  
def dfs_2(graph,x,y,start,n):
    graph[x][y] = start
    for i in graph:
        print(i)
    print()
    #오른
    if y<n:
        dfs_2(graph,x,y+1,start+1,n)
    #아래
    if x<n: 
        dfs_2(graph,x+1,y,start+1,n)


dfs(graph,x,y,1,n)
dfs_2(graph,x,y,1,n)
for i in graph[1:]:
    for j in i[1:]:
        print(j,end=' ')
    print()
```
