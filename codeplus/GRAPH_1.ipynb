```python
# dfs와 bfs

n,m,v = map(int,input().split())

graph = [[] for _ in range(n+1)]
# graph = [[], [2, 3, 4], [1, 4], [1, 4], [1, 2, 3]]
# graph = [[], [2, 3], [5, 1], [4, 1], [5, 3], [4, 2]]

for i in range(m):
    a, b = map(int, input().split())
    graph[a]+=[b]
    graph[b]+=[a]

for i in graph:
    i.sort()

visited = [0]*(n+1)
result = [[],[]]

def dfs(start, visited,graph):
    visited[start] = 1
    result[0] += [start]
    for i in graph[start]:
        if visited[i] == 0:
            dfs(i,visited, graph)

dfs(v, visited, graph)

visited = [0]*(n+1)

from collections import deque

def bfs(start, visited, graph):
    visited[start] = 1
    q = deque([start])
    while q:
        v = q.popleft()
        result[1] += [v]
        for i in graph[v]:
            if visited[i]==0:
                visited[i]=1
                q.append(i)

bfs(v, visited, graph)

for i in result:
    print(' '.join(map(str,i)))
    

```

```python
# 토마토
# 가로, 세로
from collections import deque

m,n = map(int, input().split())

graph = [list(map(int, input().split())) for _ in range(n)]

visited = [[0]*m for _ in range(n)]
# 상하좌우
dx = [1,-1,0,0]
dy = [0,0,-1,1]

def bfs(q):
    while q:
        x,y = q.popleft()
        visited[x][y] = 1
        
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if 0<=nx<n and 0 <=ny<m:
                if visited[nx][ny] == 0:
                    visited[nx][ny] = 1
                    if graph[nx][ny] == 1:
                        continue
                    if graph[nx][ny] == -1:
                        continue
                    if graph[nx][ny] == 0:
                        graph[nx][ny] = graph[x][y] + 1
                    elif graph[nx][ny] <= graph[x][y]:
                        continue
                    q.append((nx,ny))
                    
q = deque([])
for i in range(n):
    for j in range(m): 
        if graph[i][j] == 1:
            q.append((i,j))

bfs(q)

result = 0
for i in graph:
    if 0 in i:
        print(-1)
        exit()
    result = max(result, max(i))
if result == 1:
    print(0)
else:
    print(result-1) 
```

```python
# 이분 그래프
# 아이디어 깊이 탐색으로 노드에 1,2를 번갈아 표시 후
# 이미 방문했던 노드가 다시 순환될때 그룹이 같으면 no 다르면 yes
import sys
sys.setrecursionlimit(200000)
input = sys.stdin.readline

def dfs(start, visited, graph, group):
    visited[start] = group
    # print(start)
    for i in graph[start]:
        if visited[i] == 0:
            if group == 1:
                dfs(i, visited, graph, 2)
            if group == 2:
                dfs(i, visited, graph, 1)
        elif visited[i] != 0:
            if group == visited[i]:
                global result
                result = 0
                return result
            
    
k = int(input())
# 정점수, 간선수
for i in range(k):
    v, e = map(int, input().split())
    visited = [0] * (v+1)
    graph = [ [] for _ in range(v+1) ]
    for j in range(e):
        a,b = map(int, input().split())
        graph[a].append(b)
        graph[b].append(a)
    
    result = 1
    for i in range(1, v+1):
        if visited[i] == 0:
            dfs(i, visited, graph, 1)
            if result == 0:
                print('NO')
                break
    if result ==1 :
        print('YES')
    # print(graph)
    # print(visited)
```
