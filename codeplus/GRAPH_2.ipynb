```python
# Two Dots
'''아이디어: 시작점을 가지고 출발해서 시작점으로 돌아올때 4개이상의
같은 종류의 노드를 거쳤다면 싸이클이라고 판단'''

import sys 
sys.setrecursionlimit(200000)
input = sys.stdin.readline

n, m = map(int, input().split())
graph = [list(map(str, input())) for _ in range(n)]

# 상하좌우
dx = [1, -1, 0, 0]
dy = [0, 0, -1, 1]

def dfs(x,y,cnt):
    visited[x][y] = cnt
    # for k in visited:
    #     print(k)
    for i in range(4):
        nx = x+dx[i]
        ny = y+dy[i]
        if 0<=nx<n and 0<=ny<m:
            if visited[nx][ny] == 0:
                if graph[x][y]==graph[nx][ny]:
                    dfs(nx,ny,cnt+1)
            elif cnt >= 4:
                if visited[nx][ny]==1:
                    global circle
                    circle = 1
                    return circle
    
visited = [ [0]*m for _ in range(n)]

circle = 0
for i in range(n):
    for j in range(m):
        dfs(i,j,1)
        visited = [ [0]*m for _ in range(n)]
        if circle ==1:
            print('Yes')
            break
    if circle ==1:
        break
if circle ==0:
    print('No')
```

```python
# 서울 지하철 2호선

''' 아이디어: dfs로 순환선을 찾는다. 그외 노드들은 지선이다
지선은 bfs로 돌려서 가장 가까운 순환선 노드와의 거리를 찾는다'''

from collections import deque
import sys
# input = sys.stdin.readline
sys.setrecursionlimit(200000)

n = int(input())
graph = [[] for _ in range(n+1)]
visited = [0]*(n+1)

for i in range(n):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)

# print(graph)
# print(visited)

#dfs 시작: 순환선 찾기 
def dfs(start, visited, graph, node_lst):
    visited[start] = 1
    # print(node_lst)
    for i in graph[start]:
        if visited[i] == 0:
            dfs(i, visited, graph,node_lst + [i])
        elif visited[i] == 1 and i == node_lst[0]:
            if len(node_lst) >= 3:
                global stop, circled
                circled = node_lst
                stop = 1
                return node_lst
    visited[start] = 0

for i in range(1, n+1):
    stop = 0
    circled = []
    dfs(i,visited,graph,[i])
    if stop ==1:
        not_circled = []
        for j in range(1, n+1):
            if j not in circled:
                not_circled.append(j)
        # print(not_circled)
        break
    
# bfs 시작: 순환선과의 거리 찾기
def bfs(start, visited, graph):
    global cnt
    q = deque([start])
    visited[start] = 1
    while q:
        now = q.popleft()
        print(visited)
        cnt += 1
        for i in graph[now]:
            if visited[i] == 0:
                visited[i] = 1
                q.append(i)
                if i in circled:
                    return

print(not_circled)       
ans = [0]*n
for i in not_circled:
    temp = []
    print('탐색시작')
    visited = [0]*(n+1)
    bfs(i, visited, graph)
    print(len(temp))
    ans[i-1] = len(temp)

print(' '.join(map(str, ans)))
```
