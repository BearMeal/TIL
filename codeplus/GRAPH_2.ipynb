```python
# Two Dots
'''아이디어: 시작점을 가지고 출발해서 시작점으로 돌아올때 4개이상의
같은 종류의 노드를 거쳤다면 싸이클이라고 판단'''

import sys 
sys.setrecursionlimit(200000)
input = sys.stdin.readline

n, m = map(int, input().split())
graph = [list(map(str, input())) for _ in range(n)]

# 상하좌우
dx = [1, -1, 0, 0]
dy = [0, 0, -1, 1]

def dfs(x,y,cnt):
    visited[x][y] = cnt
    # for k in visited:
    #     print(k)
    for i in range(4):
        nx = x+dx[i]
        ny = y+dy[i]
        if 0<=nx<n and 0<=ny<m:
            if visited[nx][ny] == 0:
                if graph[x][y]==graph[nx][ny]:
                    dfs(nx,ny,cnt+1)
            elif cnt >= 4:
                if visited[nx][ny]==1:
                    global circle
                    circle = 1
                    return circle
    
visited = [ [0]*m for _ in range(n)]

circle = 0
for i in range(n):
    for j in range(m):
        dfs(i,j,1)
        visited = [ [0]*m for _ in range(n)]
        if circle ==1:
            print('Yes')
            break
    if circle ==1:
        break
if circle ==0:
    print('No')
```

```python
# 서울 지하철 2호선
# bfs에서 노드간의 거리를 구할때는 visited[now]+1 만큼의 cnt를하면된다

''' 아이디어: dfs로 순환선을 찾는다. 그외 노드들은 지선이다
지선은 bfs로 돌려서 가장 가까운 순환선 노드와의 거리를 찾는다'''


from collections import deque
import sys
input = sys.stdin.readline
sys.setrecursionlimit(200000)

n = int(input())
graph = [[] for _ in range(n+1)]
visited = [0]*(n+1)

for i in range(n):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)

# print(graph)
# print(visited)

#dfs 시작: 순환선 찾기 
def dfs(start, visited, graph, node_lst):
    visited[start] = 1
    # print(node_lst)
    for i in graph[start]:
        if visited[i] == 0:
            dfs(i, visited, graph,node_lst + [i])
        elif visited[i] == 1 and i == node_lst[0]:
            if len(node_lst) >= 3:
                global stop, circled
                circled = node_lst
                stop = 1
                return node_lst
    visited[start] = 0

for i in range(1, n+1):
    stop = 0
    circled = []
    dfs(i,visited,graph,[i])
    if stop ==1:
        not_circled = []
        for j in range(1, n+1):
            if j not in circled:
                not_circled.append(j)
        # print(not_circled)
        break
    
# bfs 시작: 순환선내의 노드들을 bfs탐색해서 지선과의 거리 구하기
def bfs(circled, visited, graph):
    q = deque(circled)
    for i in q:
        visited[i] = 0
    cnt = 0
    while q:
        now = q.popleft()
        for i in graph[now]:
            if visited[i] ==0 and i not in circled:
                visited[i] = visited[now]+1
                q.append(i)

bfs(circled , visited, graph)

print(' '.join(map(str, visited[1:])))

```

```python
# bfs스페셜 저지
# 아이디어: 순서를 하나씩넣으면서 bfs로 가능한지 비교하여 확인

# import sys
# input = sys.stdin.readline
# sys.setrecursionlimit(1000000)

from collections import deque

n = int(input())
graph = [[] for _ in range(n+1)]
visited = [0] * (n+1)

for i in range(n-1):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)
# print(graph)

check_order = list(map(int, input().split()))
# print(check_order)

def bfs(start, check_order, graph, visited):
    q = deque([start])
    visited[start] = 1
    idx = 1
    while q:
        if idx == n:
            print(1)
            return
        now = q.popleft()
        # print(now, q, idx)
        temp_lst = [] # 가능한 방문 뒷순서로 옮기기
        pos = 0  # 다음순서로 가능한지
        for i in graph[now]:
            if visited[i] == 0:
                if check_order[idx] == i:
                    visited[i] = 1
                    q.append(i)
                    pos = 1
                    idx +=1
                else:
                    temp_lst.append(i)
        if pos == 1:
            q.extend(temp_lst)
        elif pos == 0 and len(temp_lst) > 0:
            print(0)
            return
        idx += len(temp_lst)
                    

if len(graph[1]) ==0: 
    print(0)
else: 
    bfs(1, check_order, graph, visited)
    
```
